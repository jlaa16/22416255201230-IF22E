```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#memuat data iris
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)

#menggunakan DBSCAN 
dbscan = DBSCAN(eps=0.5, min_samples=5)
df['cluster'] = dbscan.fit_predict(df)

#menampilkan jumlah kluster yang di hasilkan (klaster -1 menandakan noise)
print("jumlah Klaster:",df['cluster'].nunique())

#visualisasi pengelompokan menggunakan scater plot
plt.figure(figsize=(8,6))
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', hue='cluster', data=df, palette='viridis', s=100)
plt.title("DBSCAN Clustering of iris Dataset")
plt.xlabes('sepal Length (cm)')
plt.ylabel('sepal Width (cm)')
plt.legend()
plt.show()
```
