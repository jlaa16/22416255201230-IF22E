```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

# Memuat data iris
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Menggunakan Agglomerative Hierarchical Clustering
agg_cluster = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')
df['cluster'] = agg_cluster.fit_predict(df)

# Visualisasi dendogram
plt.figure(figsize=(12,8))
linked = linkage(df.iloc[:,:-1], 'ward')

dendrogram(linked, orientation='top', distance_sort='descending', show_leaf_counts=True)
plt.title('Dendrogram for Hierarchical Clustering')
plt.xlabel('Data Points')
plt.ylabel('Euclidean Distance')
plt.show()
```
